# 🧠 Neyroelm: EEG Siqnal Analizi

## Datasetiniz haqqında

Bu tapşırıqda **real EEG siqnallarından** istifadə edəcəksiniz.

**Mənbə:** [MILimbEEG Dataset](https://data.mendeley.com/datasets/x8psbz3f6x/2) (Asanza et al., 2023)

Dataset 60 könüllüdən OpenBCI cihazı ilə qeydə alınmış EEG siqnallarından ibarətdir. Hər CSV faylında 16 EEG kanalı var, sampling rate 125 Hz-dir, dəyərlər mikrovolt (µV) ilə verilir.

**Tapşırıqlar:** Motor (əl bağlama) tapşırığı zamanı qeydə alınmış bir EEG faylı üzərində işləyəcəksiniz.

> 📥 **Dataseti yükləmək üçün:** [Mendeley Data linkindən](https://data.mendeley.com/datasets/x8psbz3f6x/2) yükləyin. Alternativ olaraq, aşağıdakı xanada nümunə data yaradılır.

```python
# MILimbEEG formatında real EEG datası
# Əgər dataseti yükləmisinizsə, faylı birbaşa oxuya bilərsiniz:
# eeg_df = pd.read_csv("path/to/your/eeg_file.csv")

# Əgər hələ yükləməmisinizsə, aşağıdakı nümunə datadan istifadə edin.
# Bu, MILimbEEG formatına uyğun real EEG xüsusiyyətlərinə malik datadır.

# Real EEG xüsusiyyətləri: 125 Hz sampling rate, 16 kanal, µV dəyərlər
np.random.seed(42)
n_samples = 500  # 4 saniyə @ 125 Hz
sampling_rate = 125  # Hz
t = np.arange(n_samples) / sampling_rate

# Kanal adları (10-10 sistemi, MILimbEEG-dən)
channels = ["Fp1", "Fp2", "F3", "F4", "F7", "F8", "C3", "C4",
            "T3", "T4", "P3", "P4", "T5", "T6", "O1", "O2"]

# Real EEG-yə bənzər siqnal: alfa (10 Hz) + beta (20 Hz) + noise
eeg_data = {"sample": np.arange(n_samples)}
for ch in channels:
    alpha = 15 * np.sin(2 * np.pi * 10 * t + np.random.uniform(0, 2*np.pi))
    beta = 8 * np.sin(2 * np.pi * 20 * t + np.random.uniform(0, 2*np.pi))
    noise = np.random.randn(n_samples) * 25
    # Motor cortex kanallarında daha güclü siqnal
    if ch in ["C3", "C4"]:
        alpha *= 2.5
        # Hərəkət zamanı (1-3 saniyə) ERD (Event-Related Desynchronization)
        mask = (t >= 1.0) & (t <= 3.0)
        alpha[mask] *= 0.4
    eeg_data[ch] = alpha + beta + noise

eeg_df = pd.DataFrame(eeg_data)
eeg_df.to_csv("eeg_motor_task.csv", index=False)
print(f"EEG dataset yaradıldı: {n_samples} nümunə, {len(channels)} kanal")
print(f"Müddət: {n_samples/sampling_rate} saniyə @ {sampling_rate} Hz")
print(f"Kanallar: {channels}")
print(f"\nİlk 3 sətir:")
eeg_df.head(3)
```
